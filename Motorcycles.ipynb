### 2. Convert `Price as new (MSRP)` to a number

The data is in both US dollars and euros, but luckily they're more or less 1:1 right now! We'll just convert them to a single price.

* Tip: It can't be an integer because it has missing values.

```python
#Price as new (MSRP)
df['Price as new (MSRP)'] = df['Price as new (MSRP)'].str.replace('€', '')
```

```python
df['Price as new (MSRP)'] = df['Price as new (MSRP)'].str.replace('$', '')
```

```python
df['Price as new (MSRP)'] = pd.to_numeric(df['Price as new (MSRP)'], errors='coerce').fillna(0)
```
